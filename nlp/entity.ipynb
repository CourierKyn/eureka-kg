### Cloud trems

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/604.5.6 (KHTML, like Gecko) Version/11.0.3 Safari/604.5.6'}
url='https://www.techrepublic.com/blog/the-enterprise-cloud/mini-glossary-cloud-computing-terms-you-should-know/'
response = requests.get(url=url,headers=headers)
```

```python
text=''
for i in soup.select('div.content h3'):
    try:
        text=text+i.string.strip()+'\n'
    except:
        pass
```

### Blockchain terms

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/604.5.6 (KHTML, like Gecko) Version/11.0.3 Safari/604.5.6'}
url='https://www.blockchaintechnologies.com/glossary/'
response = requests.get(url=url,headers=headers)
```

```python
html=response.text
soup = BeautifulSoup(html, 'lxml')
soup
```

```python
text=''
B1=[]
for i in soup.select('p strong'):
    try:
        print(i.get_text())
        B1.append(i.get_text().strip())
        text=text+i.get_text().strip()+'\n'
    except:
        pass
```

```python
with open('B_entity_1.txt','w',encoding='utf_8') as f:
    f.write(text)
```

```python
B0=[line.strip() for line in open('B_entity.txt').readlines()]

B='\n'.join(set(B0+B1))

with open('B_entity_all.txt','w',encoding='utf_8') as f:
    f.write(B)
```
