```python
import pandas as pd
```

```python
zh_stop_words=pd.read_csv('zh_stop_words.csv',header=None)
```

```python
zh_stop_words
```

```python
hibtc=pd.read_csv('data/out_file/hibtc.txt',sep='\t',header=None)
```

```python
hibtc
```

```python
hibtc=hibtc.set_index(0)
```

```python
stop_word=list(set(hibtc.index).intersection(set(zh_stop_words[0])))
```

```python
stop_word
```

```python
hibtc=hibtc.drop(stop_word)
```

```python
hibtc=hibtc.reset_index()
```

```python
hibtc=hibtc.drop_duplicates()
```

```python
hibtc.to_csv('data/处理后三元组/hibtc.csv',sep='\t')
```
