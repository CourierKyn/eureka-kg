### AI terms

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/604.5.6 (KHTML, like Gecko) Version/11.0.3 Safari/604.5.6'}
url='https://learn.g2crowd.com/artificial-intelligence-terms'
response = requests.get(url=url,headers=headers)
```

```python
html=response.text
soup = BeautifulSoup(html, 'lxml')
```

```python
lt=[]
j=0
while j<len(soup.select('div.cell p')):
    print(j+1)
    i=soup.select('div.cell p')[j]
    try:
        lt.append(i.get_text().strip())
        print(lt[j])
        print('-------------------------------------')
    except:
        print('error',j)
    j+=1
```

```python
data=pd.DataFrame(columns=['entity','explanation'])
j=0
for i in lt:
    try:
        temp=i.split(': ')
        data.loc[j,'entity']=temp[0]
        data.loc[j,'explanation']=temp[1]
    except:
        pass
    j+=1
```

```python
data=data.dropna()
```

```python
data.to_csv('A_entity_explanation.csv',index=0,encoding='utf_8')
```
