```python
import re
import json
from pandas.core.frame import DataFrame
```

```python
patternOfLeading_50 = re.compile(r'#\d\d (.+)')   
```

```python
text = open(r"texe_10_11.txt",'r',encoding='utf-8').read()
Leading_50 = patternOfLeading_50.findall(text)
```

```python
Leading_50
```

```python
len(Leading_50)
```

```python
patternOfEmerging_50 = re.compile(r'\ue801 ([\s\S]+?)(?=\n[^a-zA-z])')
```

```python
Emerging_50 = patternOfEmerging_50.findall(text)
```

```python
Emerging_50
```

```python
len(Emerging_50)
```

```python
Emerging_50[4] = 'Blackmoon Financial Group'
```

```python
Company100 = DataFrame({'Leading_50':Leading_50,'Emerging_50':Emerging_50})
```

```python
Company100
```

```python
with open('Company100.json', 'w') as f:
        json.dump({'Leading_50':Leading_50,'Emerging_50':Emerging_50}, f, indent=4)
```
