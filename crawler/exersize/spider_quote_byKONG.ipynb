```python
from bs4 import BeautifulSoup
import requests
import re
import os
import pandas as pd
from pandas import Series,DataFrame
```

```python
URL = "http://quotes.toscrape.com/"
html = requests.get(URL).text
soup = BeautifulSoup(html, 'lxml')
```

```python
text = [i.get_text()[1:-1] for i in soup.select("span.text")]
author = [i.get_text() for i in soup.select("small")]
tag = [i['content'] for i in soup.select("div > meta")]
```

```python
df = DataFrame({'text':text,'author':author,'tag':tag})
```
