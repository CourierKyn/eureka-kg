```python
lst = []
for i in data:
    if i['response']['search']:
        i['response']['search'] = [i['response']['search'][0]]
        if not i.get('mark'):
            if i['entity'].lower() == i['response']['search'][0]['match']['text'].lower():
                lst.append(i)
                cnt += 1
#             elif :
#                 display(i)
            else:
                inf = True
                for j in data:
                    if j['entity'] == i['entity'] and j.get('mark'):
                        j['response']['search'] = [j['response']['search'][0]]
                        if i['response']['search'][0]['label'].lower() == j['response']['search'][0]['label'].lower():
                            lst.append(i)
                            lst.append(j)
                            inf = False
                            break
                if inf and ('(' in i['entity'] or ' / ' in i['entity']) and i['response']['searchinfo']['search'].lower() == i['response']['search'][0]['match']['text'].lower():
                    if i['entity'] != 'Interpretation (logic)':
                        lst.append(i)
```
