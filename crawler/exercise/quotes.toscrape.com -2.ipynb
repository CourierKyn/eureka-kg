```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
```

```python
def get_soup(url):
    html = requests.get(url).text
    soup = BeautifulSoup(html, 'lxml')
    return soup
```

```python
def get_text(sub_soup):
    text = [i.get_text()[1:-1] for i in sub_soup.select("span.text")]
    if text == []:
        text.append(np.nan)
    
    return text[0]
```

```python
def get_author(sub_soup):
    author =  [i.get_text() for i in sub_soup.select("small")]
    if author == []:
        author.append(np.nan)
    return author[0]
```

```python
def get_tag(sub_soup):
    tag = [i['content'] for i in sub_soup.select("div > meta")]
    if tag == []:
        tag.append(np.nan)
    return tag
```

```python
def get_next_pages_beautifulSoup(sub_soup, start_url):
    try:
        url = start_url + sub_soup.select('li.next')[0].select('a')[0].get('href')
        print(url)
        html = requests.get(url).text
        soup = BeautifulSoup(html, "lxml")
        return soup
    except IndexError:
        print('no more page')
        return 0
```

```python
def combine_information(text, author, tag, dafr):
    df = pd.concat([dafr,DataFrame({'text':text,'author':author,'tag':tag})],ignore_index=True)
    return df
```

```python
def get_block_information(sub_soup, df):
    for i in sub_soup.select("div.quote"):
        text = get_text(i)
        tag = get_tag(i)
        author = get_author(i)
        
        df = combine_information(text, author, tag, df)
    return df
```

```python
def main():
    url = "http://quotes.toscrape.com"
    soup = get_soup(url)
    df = DataFrame()
    df = get_block_information(soup, df)
    while get_next_pages_beautifulSoup(soup, url) != 0:
        soup = get_next_pages_beautifulSoup(soup, url)
        df = get_block_information(soup, df)    
    print(df)
    return
```

```python
main()
```
